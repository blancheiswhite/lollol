```python
from PIL import Image
```

```python
newimage = Image.open('waysofseeing2.png')

def flpp(pngimage):

    width = pngimage.size[0]
    
    if width % 2 == 0:
        width_half = width / 2
    else:
        width_half = (width - 1) / 2     
        for y in range(pngimage.size[1]):
            for x in range(0, width_half):
                right = pngimage.getpixel((width-x-1,y))
                left = pngimage.getpixel((x,y))
                pngimage.putpixel((x,y), left)
                pngimage.putpixel((width-x-1,y), right)
```

```python
flpp(newimage)
newimage
```

```python
newimage = Image.open('waysofseeing2.png')

def flap(pngimage):

    width = pngimage.size[0]
    
    if width % 2 == 0:
        width_half = width // 2
    else:
        width_half = (width - 1) // 2     
    for y in range(pngimage.size[1]):
        for x in range(0, width_half): 
            left = pngimage.getpixel((x,y))
            right = pngimage.getpixel((width-x-1,y))
            pngimage.putpixel((width-x-1,y), left)
            pngimage.putpixel((x,y), right)
                
```

```python
flap(newimage)
newimage.save('test3.png')
```
